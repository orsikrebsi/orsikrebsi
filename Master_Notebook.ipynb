### Possible Issue - Token Expiration

If for any reason (accidental deletion, incorrectly set expiration date, etc.), your github token expires after you've already cloned your repo, git won't let you pull or push any changes from or onto the remote repo.

Don't worry, you don't need to delete and reclone your repo and possibly lose any changes. Instead, simply run the following command, placing your new token at `<new-github-token>` in the path.

````python
!git -C {drive_repo} remote set-url origin https://<new-github-token>@github.com/{repo}
````

## Setting up local editable package

### What is an editable package

`pip` allows you to install a local package in editable form using the command
```bash
pip install -e /path/to/your/package
```

Doing this allows python to use the package directly from that local package, meaning local changes to the package update the behavior of said package without the need to reinstall.

> Note: While you don't need to reinstall your editable package each time you edit its code, you will have to restart your kernel

````python
!pip install -e {drive_repo}/src/
````
